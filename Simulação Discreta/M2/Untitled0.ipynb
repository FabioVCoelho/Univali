```
import random as rd
import math
import matplotlib.pyplot as plt
```

```
def uniform(size,min,max):
  return [min+(max-min)*rd.random() for i in range(size)]
```

```
def triangular(size,min,max,moda):
  result = []
  var = (moda-min)/(max-min)
  for i in range(size):
    R = rd.random()
    if R < var:
      x = min + ((R*(moda-min)*(max-min))**(1/2))
    else:
      x = max - (((1-R)*(max-moda)*(max-min))**(1/2))
    result.append(x)
  return result
```

```
def exponential(size, medium_value):
  return [(-medium_value)*(math.log(rd.random())) for i in range(size)]
```

```
def normal(size, medium_value, variance):
  result = []
  for i in range(size):
    R1 = rd.random()
    R2 = rd.random()
    z = ((-2*math.log(R1)) ** (1/2)) * (math.sin(2*math.pi*R2))
    x = (variance + (medium_value*z))
    result.append(x)
  return result
```

```
def distribuicoes(escolha):
  if escolha == 1:
    a = float(input("Escolha o Valor A: "))
    b = float(input("Escolha o Valor B: "))
    quantidade = int(input("Escolha a quantidade de valores a ser gerado: "))
    uniform_list = uniform(quantidade,a,b)
    return uniform_list

  elif escolha == 2:
    min = float(input("Escolha o Valor min: "))
    max = float(input("Escolha o Valor max: "))
    moda = float(input("Escolha o Valor moda: "))
    quantidade = int(input("Escolha a quantidade de valores a ser gerado: "))
    triangular_list = triangular(quantidade,min,max,moda)
    return triangular_list

  elif escolha == 3:
    u = float(input("Escolha o Valor Medio: "))
    quantidade = int(input("Escolha a quantidade de valores a ser gerado: "))
    exponential_list = exponential(quantidade,u)
    return exponential_list

  elif escolha == 4:
    u = float(input("Escolha o Valor Medio: "))
    o = float(input("Escolha o Valor da Variancia: "))
    quantidade = int(input("Escolha a quantidade de valores a ser gerado: "))
    normal_list = normal(quantidade,u,o)
    return normal_list

  elif escolha == 5:
    quantidade = int(input("Insira os valores separados por , e números decimais representados com ."))
    manual_list = quantidade.split(",")
    return manual_list
  else:
    return "Não foi escolhido nenhuma opção válida"
```

```
# Tempo entre chegadas
# Pode ser aleatório ou deterministico
print("Qual a distribuição que será usada para o Tempo entre chegadas (TEC):")
print("1 - Uniforme")
print("2 - Triangular")
print("3 - Exponencial")
print("4 - Normal")
print("5 - Manual")
escolha = int(input())
tec_distribution = distribuicoes(escolha)
print(tec_distribution[:10])
```

```
# Tempo de serviço
# Pode ser aleatório ou deterministico
print("Qual a distribuição que será usada para o Tempo de serviço (TS):")
print("1 - Uniforme")
print("2 - Triangular")
print("3 - Exponencial")
print("4 - Normal")
print("5 - Manual")
escolha = int(input())
ts_distribution = distribuicoes(escolha)
print(ts_distribution[:10])
```

```
def limites(escolha):
  print(escolha)
  if escolha == 1:
    print("Com quantos segundos deverá finalizar a simulação?")
    return int(input())
  elif escolha == 2:
    print("Com quantos carros deverá finalizar a simulação?")
    return int(input())
  
```

```
print("Qual variável será utilizada para finalizar a simulação: ")
print("1 - Tempo")
print("2 - Quantidade de carros")
escolha_limite = int(input())
if escolha_limite == 1:
  tempo_limite = limites(escolha_limite)
  carro_limite = math.inf
else:
  carro_limite = limites(escolha_limite)
  tempo_limite = math.inf
```

```
tempo_relogio = []
tempo_inicio_servico = []
tempo_cliente_na_fila = []
tempo_final_do_servico = []
tempo_cliente_no_sistema = []
tempo_livre_do_operador = []
esperou_na_fila = []
index = 0

def run_simulation():
  index = 0
  value = tec_distribution[0]
  tempo_relogio.append(value)
  tempo_inicio_servico.append(ts_distribution[index])
  tempo_cliente_na_fila.append(tempo_inicio_servico[index] - tempo_relogio[index])
  tempo_final_do_servico.append(tempo_inicio_servico[index] + ts_distribution[index])
  tempo_cliente_no_sistema.append(ts_distribution[index] + tempo_cliente_na_fila[index])
  tempo_livre_do_operador.append(tempo_relogio[index])
  if (tempo_cliente_na_fila[index] > 0):
    esperou_na_fila.append(1)
  else:
    esperou_na_fila.append(0)

  print("Valor TEC: " + str(value))
  print("Tempo do relogio: " + str(tempo_relogio[index]))
  print("Valor TS: " + str(ts_distribution[index]))
  print("Tempo inicio serviço: " + str(tempo_inicio_servico[index]))
  print("Tempo cliente na fila: " + str(tempo_cliente_na_fila[index]))
  print("Tempo final do serviço: " + str(tempo_final_do_servico[index]))
  print("Tempo do cliente no sistema: " + str(tempo_cliente_no_sistema[index]))
  print("Tempo livre do operador: " + str(tempo_livre_do_operador[index]))
  print("Cliente esperou na fila: " + str(esperou_na_fila[index]))

  while (tempo_relogio[-1] < tempo_limite and index < carro_limite):
    index = index + 1
    value = tec_distribution[index]
    tempo_relogio.append(tempo_relogio[index - 1] + value)
    tempo_inicio_servico.append(max(tempo_final_do_servico[index - 1], tempo_relogio[index]))
    tempo_cliente_na_fila.append(tempo_inicio_servico[index] - tempo_relogio[index])
    tempo_final_do_servico.append(tempo_inicio_servico[index] + ts_distribution[index])
    tempo_cliente_no_sistema.append(ts_distribution[index] + tempo_cliente_na_fila[index])
    tempo_livre_do_operador.append(tempo_inicio_servico[index] - tempo_final_do_servico[index - 1])
    if (tempo_cliente_na_fila[index] > 0):
      esperou_na_fila.append(1)
    else:
      esperou_na_fila.append(0)

    print("Index: " + str(index))
    print("Valor TEC: " + str(value))
    print("Tempo do relogio: " + str(tempo_relogio[index]))
    print("Valor TS: " + str(ts_distribution[index]))
    print("Tempo inicio serviço: " + str(tempo_inicio_servico[index]))
    print("Tempo cliente na fila: " + str(tempo_cliente_na_fila[index]))
    print("Tempo final do serviço: " + str(tempo_final_do_servico[index]))
    print("Tempo do cliente no sistema: " + str(tempo_cliente_no_sistema[index]))
    print("Tempo livre do operador: " + str(tempo_livre_do_operador[index]))
    print("Cliente esperou na fila: " + str(esperou_na_fila[index]))
```

```
run_simulation()
```
